```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv("data/endurance/1bpc_bwa_0_cb_16att_1024cells_endurance.csv", delimiter="\t", names=["t", "successes", "failures", "reads", "sets", "resets", "cycles", "read_bits", "set_bits", "reset_bits", ])
data.plot("cycles", "successes")
data.plot("cycles", "failures")
plt.show()
plt.plot(data["successes"]/1024, data["successes"].diff()/(data["cycles"].diff()*10e-9))
plt.plot(data["successes"]/1024, data["failures"].diff()/(data["cycles"].diff()*10e-9))
```
